```python
my_bar_grouped(data, 'purpose', 'task', open_text['purpose'])
```

```python
data['task_group'] = ''
data.loc[(data.task == 1) | (data.task == 2), 'task_group'] = 'low'
data.loc[(data.task == 3) | (data.task == 4), 'task_group'] = 'medium'
data.loc[(data.task == 5) | (data.task == 6), 'task_group'] = 'high'
```

```python
fig, ax = plt.subplots(figsize=(9.2, 5))
x = data.groupby(['purpose', 'task_group']).size().reset_index()
ax.invert_yaxis()
ax.xaxis.set_visible(False)
ax.set_xlim(0, 240)
sums = [0, 0, 0]
colors = plt.colormaps['pink'](np.linspace(0.08, 0.35, 3))
legend = ['Wrong answer', 'Partially correct answer', 'Correct answer']
for index, i in enumerate([0, 0.5, 1]):
    curr_sum = [
        int(x[(x.purpose == i) & (x.task_group == 'low')][0]),
        int(x[(x.purpose == i) & (x.task_group == 'medium')][0]),
        int(x[(x.purpose == i) & (x.task_group == 'high')][0])
    ]
    r = ax.barh(
        ['low', 'medium', 'high'],
        curr_sum,
        left=sums,
        height=0.5, 
        label=legend[index],
        color=colors[index],
        linewidth=0
    )
    ax.bar_label(r, label_type='center', color='white', labels=map(lambda x: str(x), curr_sum))
    ax.legend(ncol=len(legend), bbox_to_anchor=(0, 1),
              loc='lower left')
    for j, v in enumerate(sums):
        sums[j] += curr_sum[j]
ax.set_ylabel('Visual complexity')
ax.set_xlabel('Number of responses')
plt.show()
```
